## agg() or aggregate()
- It is mostly used in group purpose
- Aggregate using one or more operations over the specified axis.

```python
# Dataframe
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
```

```python
data = pd.read_csv("C:\\Users\\hp\\Downloads\\archive\\train.csv")
```

```python
data.isna().sum()
```

```python
def get_initials(value):
    name = value.split(',')[1]
    initials = name.split('.')[0]
    return initials.strip()
```

```python
data['Initials'] = titanic_df['Name'].apply(get_initials)
```

```python
data["Initials"].unique()
```

```python
data['Initials'].replace(
    to_replace={'Ms':'Miss', 'Capt':'Mr', 'Dr':'Mr', 'Col':'Mr', 'Mme':'Mrs', 
               'Don':'Mr', 'Rev':'Mr', 'Major':'Mr', 'Mlle':'Miss', 'the Countess':'Mrs', 
               'Jonkheer':'Master', 'Lady': 'Mrs', 'Sir':'Mr'}, 
    inplace=True)
```

```python
data['Initials'].value_counts().plot.bar()
```

```python
grouped = data.groupby(by='Initials')
```

```python
grouped['Age'].agg(NaN=lambda x: x.isna().sum(), NotNaN=lambda x: x.notna().sum())
```

```python
grouped['Age'].agg(['max', 'min', 'mean'])
```
