```python
import numpy as np
import scipy 
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = np.random.standard_t(10, size=1000)
x[:15]
```

## Measure of Dispersion/Variability

A measure of dispersion, also known as a measure of variability or spread, is a statistical metric that quantifies how the data points in a dataset are spread out or dispersed around a central value. It provides insight into the distribution and variability of the data.

Measures of dispersion help us understand the range or extent to which the data points deviate from a central tendency. They provide additional information beyond the measures of central tendency (such as mean, median, and mode) by capturing the spread, scatter, or variability within the dataset.

Some of the common measures of dispersion are:
- Variance
- Standard Deviation
- Range
- Interquartile Range (IQR)
- Mean Absolute Deviation (MAD)

### Variation
The variance, denoted by $\sigma^2$ , is another measure of spread of a dataset. It is the average of the squared deviations from the mean. Variance provides a measure of how the individual data points differ from the mean.

```python
variance = np.var(x)
print("Variation of x:", variance)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=np.mean(x), color='yellow')
plt.axvline(x = variance, color= 'red')
```

This plot show us how scatter the data is from mean. By yellow line we find out mean and by red line we know it is central tendency.

### Standard Deviation
The standard deviation, denoted by $\sigma$ 
, measures the dispersion or spread of a dataset. It quantifies how much the values deviate from the mean. A higher standard deviation indicates greater variability in the data.

```python
standard_deviation = np.std(x)
print(standard_deviation)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=np.mean(x), color='yellow')
plt.axvline(x = standard_deviation, color= 'red')
```

<br>

### Range
The range is a measure of dispersion that indicates the difference between the maximum and 
minimum values in a dataset. It provides a simple representation of the spread of the data.

```python
rangee = np.max(x) - np.min(x)
print("Range of x:", rangee)
```

<br>

### Interquartile  Range
The interquartile range is a measure of dispersion that captures the spread of the middle 50% of the data, excluding the extreme values. It is calculated as the difference between the upper quartile (Q3) and the lower quartile (Q1).

```python
q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)

iqr = q3- q1

print("Interquartile range of x:", iqr)
```

```python
max_boundary = q3 + (1.5 * iqr)
min_boundary = q1 - (1.5 * iqr)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=np.median(x), color='yellow')
plt.axvline(x = q1, color= 'red')
plt.axvline(x = q3, color= 'red')
plt.axvline(x = max_boundary, color= 'red')
plt.axvline(x = min_boundary, color= 'red')
```

```python
plt.boxplot(x, vert=False,)
```

<br>

### Mean Absolute Deviation(MAD)
The mean absolute deviation is a measure of dispersion that quantifies the average absolute difference between each data point and the mean. It represents the average distance of each data point from the central value.

```python
mad = (1/len(x)) * sum(np.abs(x-np.mean(x)))
print(mad)
```

```python
#Alternative
mad = np.mean(np.abs(x - np.mean(x)))
print("Mean Absolute Deviation of x:", mad)
```

```python
plt.hist(x, color='blue')
plt.axvline(x=np.mean(x), color='yellow')
plt.axvline(x = standard_deviation, color= 'red')
plt.axvline(x = mad, color = 'cyan')
```
