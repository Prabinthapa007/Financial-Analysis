## Central Limit Theorem

To understand Central Limit Theorem we should know two measure distributions:
- Normal Distribution
- Sampling from a Statistical Distribution

**Statistical Sampling Distribution** is a type of probability distribution created by drawing many random samples of a given size from the same population.
- Suppose you draw 10 apples as a random sample. Then you calculate the mean of that sample. This is one sample mean from one sample.
- Likewise you draw 50 time 10 apples as a random sample. Then you will have 50 sample mean from 50 sample.
- Then you plot this to see distribution of sample mean. This is statistical sampling distribution.

Now, **Central Limit Theorem** says that the sampling distribution of the mean will always be normally distributed as long as the sample size is large enough, independent, and  identical distribution.

### Assumptions of the Central Limit Theorem

Before we dive into the details, let's outline the assumptions of the Central Limit Theorem:
- **Independence**: The random variables being sampled must be independent of each other.
- **Identical Distribution**: The random variables must have the same distribution, with the same mean and standard deviation.
- **Sample Size**: The sample size should be sufficiently large. The CLT usually works well for sample sizes greater than 30, although there are cases where it can still provide reasonably accurate results for smaller sample sizes.

## Demonstrating CLT

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Uniform Distribution

```python
np.random.seed(99)
population = np.random.uniform(0,1, size=100000)
plt.figure(figsize=(9, 5))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Uniform)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
population.shape
```

```python
sample_size =(5, 10, 30, 100, 500, 1200)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)
    

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### Pareto Distribution
**Pareto Distribution**: Named after the Italian economist Vilfredo Pareto, the Pareto distribution is a power-law probability distribution that describes the distribution of wealth in a society, among other phenomena. It is characterized by a heavy right tail, indicating that a small number of observations have a disproportionately large impact.

```python
np.random.seed(32)
population = np.random.pareto(a=50, size=100000)
plt.figure(figsize=(6, 3))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Pareto)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size =(5, 10, 30, 100, 500, 1000)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### Cauchy Distribution
The **Cauchy distribution**, named after the mathematician Augustin-Louis Cauchy, is a continuous probability distribution. It is notable for its characteristic "heavy tails," which means that it has a higher probability of producing extreme outliers compared to other distributions like the normal distribution.

```python
np.random.seed(32)
population = np.random.standard_cauchy( size=100000)
plt.figure(figsize=(6, 3))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Cauchy)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size =(5, 10, 30, 100, 500, 10000)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```
