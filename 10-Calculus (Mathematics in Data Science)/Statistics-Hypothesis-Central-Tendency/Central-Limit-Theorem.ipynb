### Demonstrating CLT

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Uniform Distribution

```python
np.random.seed(99)
population = np.random.uniform(0,1, size=100000)
plt.figure(figsize=(9, 5))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Uniform)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
population.shape
```

```python
sample_size =(5, 10, 30, 100, 500, 1200)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)
    

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### Exponential Distribution

```python
np.random.seed(32)
population = np.random.exponential(1, size=100000)
plt.figure(figsize=(6, 3))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Exponential)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size =(5, 10, 30, 100, 500, 1000)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

### Poisson Distribution

```python
np.random.seed(32)
population = np.random.poisson(lam=300.0, size=100000)
plt.figure(figsize=(6, 3))
plt.hist(population, bins=50, density=True, color='blue', alpha=0.8)
plt.title("Population Distribution (Poisson)", color='Red', fontsize=15)
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sample_size =(5, 10, 30, 100, 500, 1000)
sample_means = [ ]
for n in sample_size:
    samples = np.random.choice(population, size=(1000, n), replace = True)
    means = np.mean(samples, axis=1)
    sample_means.append(means)

for i, means in enumerate(sample_means):
    plt.subplot(2, 3, i+1)
    plt.hist(means, bins=30, density=True, color='green', alpha=0.6)
    plt.title(f"Sample Size: {sample_size[i]}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
plt.tight_layout()
plt.show()
```
