```python
import numpy as np
import scipy 
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = np.random.standard_t(10, size=1000)
x[:15]
```

## Measure of Central Tendency

Neasure of central tendency are statistical measures that provides information about the central or typical value of a dataset. They give us a sense of where the data is cnetered or concentrated. The three common measures of central tendency are:
- Mean
- Median
- Mode

### Mean
The mean, denoted by   , is a measure of central tendency that represents the average value of a dataset. It is calculated by summing all the values in the dataset by dividing by the number of observations.



```python
_mean = np.mean(x)
print(f"Mean of x: {_mean}")
```

```python
plt.hist(x, color='blue')
plt.axvline(x= _mean, color ='yellow')
```

- `plt.hist` give histogram plot of a mean of x
- `plt.axvline` gives the line where the concentrated data are more. In other word, `plt.axvline` gives line where the value has highest occurence.

<br>

### Median

```python
_median = np.median(x)
print(f"Median of x: {_median}")
```

```python
plt.hist(x, color = 'blue')
plt.axvline(x = _median, color = 'red')
```

```python
#### Lets put mean and median together in a figure.
plt.hist(x, color='blue')
plt.axvline(x=_mean, color='yellow')
plt.axvline(x=_median, color='red')
```

<br>

### Mode
The mode is a measure of central tendency that represents the most frequently occurring values in a dataset. A dataset can have one mode(unimodal) or multiple modes (multimode).

```python
_mode = stats.mode(x).mode
print(f"Mode value of x: {_mode}")
```

```python
plt.hist(x, color='blue')
plt.axvline(x=_mean, color='yellow')
plt.axvline(x=_median, color='red')
plt.axvline(x=_mode, color='green')
```
