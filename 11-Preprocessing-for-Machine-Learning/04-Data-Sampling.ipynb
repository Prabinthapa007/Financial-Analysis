### 2. Stratified Sampling
Stratified method is a type of probability sampling where the population is first divided into subgroup as according to thier characteristics (or strata) who all share a similar charactersitics. It is used when we might reasonably expert the measurement of interest to vary between the different subgroups, and we want to ensure representation, from all the subgroupbs.

```python
# Pandas Way
df = pd.DataFrame(
    {
        'student_id' : np.arange(1, 101),
        'class' : np.random.choice(['A', 'B', 'C', 'D'], size=100),
        'grade' : np.random.randint(0, 101, size=100)
    }
)
df.head()
```

```python
sample_func = lambda x: x.sample(10)

sampled_data = df.groupby('class').apply(sample_func)\

sampled_data = sampled_data.reset_index(drop=True)
sampled_data
```
