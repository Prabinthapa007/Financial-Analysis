```python
import pandas as pd
```

## Creating DataFrames

```python
# Defining singular data for dataframes 
# Using dictionary 
data = {
    "name": "Bob",
    "age": 25
}
df = pd.DataFrame(data, index=[1])
print(df)
```

```python
# Dictionary of two names and age
data = {
    "name": ["Bob","Charlie", "Ram"],
    "age": [25, 22, 24]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Using list for creating data
data = [["Bob", 25], ["Charlie", 22]]
df = pd.DataFrame(data)
print(df)  #Here name of column is not defined we should give name.
```

```python
data = [["Bob", 25], ["Charlie", 22]]
df = pd.DataFrame(data, columns=["Name", "age"])
print(df)
```
