### What is Heatmap Brief about Heatmap Plot. Why is it necessary and examples.

- `Heat Maps` actually are graphical representations of data that utilize color-coded systems. It locate visually the volume of events within a dataset.
- `Heat Maps` are multifaceted and can be used to show density of population on a map.

- It is necessary because they are extemely `versatile` and `efficient` in drawing attention to trends. It is `self-explanatory`. The darker the shade, the greater the quantity (the higher the value, the tighter the `dispersion`, etc.).

### DataFrame.plot.hexbin()
The `DataFrame.plot.hexbin()` function in pandas is used to create a hexagonal binning plot. This type of plot is useful for visualizing the distribution of two numerical variables when there are a large number of data points. It divides the plot area into hexagonal bins and colors each bin based on the number of points it contains.

```python
data.plot.hexbin(x="Age", y="Fare", title="HexBin ",gridsize=30, cmap='coolwarm')
```

<br><br>

### DataFrame.background_gradient()
The `DataFrame.background_gradient()` method in pandas is used to apply a `gradient` color scheme to the `background` of a `DataFrame` based on the `cell values`. This can be useful for visually highlighting patterns or trends in the data. 

```python
corr = data[["Survived", "Pclass", "Age", "Fare"]].corr()
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

#### Example usage of heat map using matplotlib and seaborn with different datasets.

```python
Index = [0, 1, 2, 3, 4, 5]
col = ['A', 'B', 'C', 'D', 'E']
df = pd.DataFrame(abs(np.random.randn(6, 5)), index= Index, columns= col)

print(df)
print("+"*70)
#Using heatmap with the help of seaborn library
vis1 = sns.heatmap(data=df, annot=True, fmt='.2f')
plt.show()
plt.close()
```
